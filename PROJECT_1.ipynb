# EXPLORATORY DATA ANALYSIS

## Top Selling Games:

```python
top_selling = df2.groupby("Game")["Global"].sum().reset_index().sort_values(by = "Global",ascending = False).head(10)
plt.figure(figsize = (10,3))
sns.set_style("darkgrid")
sns.barplot(x = "Global", y = "Game", data = top_selling, hue = "Game", palette = "autumn")
plt.title("TOP GLOBALLY SELLING GAMES")
plt.xlabel("Total Sales (millions)")
plt.ylabel("Games")
sns.despine()
plt.show()
```

## Sales By Platform:

```python
top_publisher = df2.groupby("Publisher")["Global"].sum().reset_index().sort_values(by = "Global", ascending = False).head(10)
plt.figure(figsize = (10,3))
sns.barplot(x = "Global", y = "Publisher", data = top_publisher, hue = "Publisher", palette = "viridis")
plt.title('TOTAL SALES BY GAME PUBLISHERS')
plt.xlabel('Total Sales (millions)')
plt.ylabel('Publishers')
plt.grid()
plt.show()
```

## Popular Genres:

```python
top_genre = df2.groupby("Genre")["Global"].sum().reset_index().sort_values(by = "Global", ascending = False).head(10)
plt.figure(figsize = (10,3))
sns.lineplot(x = "Global", y = "Genre", data = top_genre, color = "r")
plt.title("MOST POPULAR GENRE")
plt.xlabel("Total Sales(millions)")
plt.ylabel("Genre")
plt.show()
```

## Popular Genre Over the Year:

```python
df2 = df2.head(20)
plt.figure(figsize = (10,4))
sns.set_style("darkgrid")
sns.lineplot(x = "Year",y="Global", data = df2, hue = "Genre" )
plt.title("TOP GENRE OVER THE YEAR")
plt.xlabel("Year(2013 - 2020)",color = "r")
plt.ylabel("Global_Sales",color = "r")
plt.show()
```

## Game Sales by Region Over the Year(2013-2020):

```python
df_new = df2.melt(
    id_vars = ["Year","Game","Global"],
    value_vars = ["North America","Europe","Japan","Rest of World"],
    var_name = "Region",
    value_name = "Sales"
)
fg = sns.FacetGrid(df_new,col= "Region", hue= "Region")
fg.map(plt.bar,"Year","Sales",edgecolor = "None").add_legend()
sns.set_context("notebook",font_scale = 1)
plt.show()
```
