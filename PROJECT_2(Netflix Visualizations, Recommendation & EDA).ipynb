## Data Transformation:

### üßπ Data Normalization for Multi-Valued Fields

To make the dataset more analysis-friendly, multi-valued fields were split and normalized using the following transformations:

---

#### 1. Splitting `listed_in` Column:
- Used `str.split(", ")` to separate multiple categories in the `listed_in` column.  
- Applied `explode()` to create separate rows for each `category`.  
- Reset the index using `reset_index(drop=True)` to keep the DataFrame tidy.

---

#### 2. Splitting `country` Column:
- Split the `country` column where multiple countries were listed using `str.split(", ")`.  
- Used `explode()` to assign one `country` per row.  
- Removed empty entries using `file = file[file["country"] != ""]`.

---

These transformations improve `granularity` and enable more accurate `filtering`, `grouping`, and `analysis` of `genres` and `countries`.


```python
file["listed_in"] = file["listed_in"].str.split(", ")
file = file.explode("listed_in").reset_index(drop=True)
file
```

```python
file["country"] = file["country"].str.split(", ")
file = file.explode("country").reset_index(drop = True)
file = file[file["country"] != ""]
file
```

```python
du = file['rating'].value_counts()
du_values = du[du > 0].index.tolist()
print(du_values)
```

```python
rating_meanings = {
    # TV Ratings
    "TV-Y": "All children (ages 2-6)",
    "TV-Y7": "Older children (7+)",
    "TV-Y7-FV": "Older children with fantasy violence",
    "TV-G": "General audience (all ages)",
    "TV-PG": "Parental guidance suggested",
    "TV-14": "Parents strongly cautioned (under 14)",
    "TV-MA": "Mature audience only (17+)",

    # Movie Ratings
    "G": "General audiences (all ages)",
    "PG": "Parental guidance suggested",
    "PG-13": "Parents strongly cautioned (under 13)",
    "R": "Restricted (under 17 with adult)",
    "NC-17": "No one 17 and under admitted",
    
    # Others
    "NR": "Not Rated (content unknown)",
    "UR": "Unrated (content unknown)"
}
```

```python
file["rating"] = file["rating"].replace(rating_meanings, regex = True)
file
```

## Datetime Preprocessing:

-> Converted the date_added column to datetime using the format `%d-%b-%y`.<br>
-> Used errors='coerce' to handle invalid formats by setting them to `NaT`.<br>
-> Extracted only the month name (e.g., "January") from the valid dates.<br>
-> Removed rows with missing date_added values using `dropna()` for clean data.<br>

```python
file["date_added"] = pd.to_datetime(file["date_added"], format = "%d-%b-%y", errors = "coerce")
file["date_added"] = file["date_added"].dt.strftime("%B")
file
```

```python
file ["date_added"].isnull().sum()
```

```python
file = file.dropna(subset = ["date_added"])
file.isnull().sum()
```

# EXPLORATORY DATA ANALYSIS

## Data Visulaisation:

### üìä Data Visualization Overview

This section presents various insights on `Movies` and `TV Shows` using multiple visual analyses:

- **Analysis of Movies vs TV Shows:**  
  A comparison of the number of `movies` and `TV shows` released, helping to identify which format dominates the platform.

- **Optimal Month for Content Release:**  
  Identifies the `month with the least content additions`, helping `producers` choose the best time to release new content with `lower competition`.

- **Movie Ratings Analysis:**  
  Shows the distribution of `movie ratings`, offering insight into the `quality` and `audience reception` of films.

- **Most Popular Genres:**  
  Highlights the most common `genres` in both `movies` and `TV shows`, showing `viewer preferences`.

- **Analysis of Movie Duration:**  
  Examines the `length` of movies, revealing trends in `average duration` and variation across content.

- **TV Shows with the Most Seasons:**  
  Lists TV shows with the `highest number of seasons`, showcasing `long-running` and possibly `popular` series.

- **Most Content Creating Countries:**</br>
  Lists the `Top 10 countries` that creating the largest numbe of `content`.
  

## Analysis of Movies v/s TV Shows:

```python
plt.figure(figsize = (12,4))
sns.countplot(x = "type",data = file, hue = "type", palette = "Set2", saturation = 1)
sns.set_style("darkgrid")
plt.title("Analysis of Movies vs TV Shows", color = "r",fontsize = 15)
plt.xlabel("MOVIE V/S TV SHOW", color = "#003366", fontsize = 12)
plt.ylabel("CONTENT TYPE" ,color = "#003366", fontsize = 12)
plt.show()
```

- It clearly shows that `Movies` significantly outnumber `TV Shows`.
- The `Movie` count exceeds `14,000`, while the `TV Show` count is much lower.
- This indicates that the platform prioritizes or hosts more `movie` content compared to `TV shows`.

## If a Producer Wants to Release Some Content, which Month must he do so?
( Month when least amount of content is added):

```python
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
plt.figure(figsize = (12,4))
sns.countplot( x = "date_added", data = file , order = month_order, hue = "type", palette = "viridis")
plt.title("NETFLIX CONTENTS UPDATE",color = "r", fontsize = 20)
plt.xlabel("MONTHS", color = "#003366", fontsize = 15)
plt.ylabel("NO. OF COUNTS", color = "#003366", fontsize = 15)
plt.show()
```

- It is evident that `Movies` are added in much higher quantities compared to `TV Shows` throughout the year.
- Months like `January`, `July`, and `October` show the highest content updates, while `February` and `May` reflect relatively lower counts.

## Movie Ratings Analysis:

```python
plt.figure(figsize = (12,6))
sns.countplot(y = "rating",data = file, hue = "type")
plt.title("ANALYSIS",color = "r", fontsize = 20)
plt.xlabel("NO. OF COUNTS",color = "#003366", fontsize = 15)
plt.ylabel("RATING OF MOVIES & TV SHOWS",color = "#003366", fontsize = 15)
plt.grid()
plt.show()
```

- The highest number of contents fall under the rating `Mature audience only (17+)`, indicating a large portion of adult-focused content.

## Most Popular Genres In (Movies & TV Shows):

```python
# Arranging the data for movie
movie_d = file[file["type"] == "Movie"]
listed_in_series = pd.Series(movie_d["listed_in"])
category_counts = listed_in_series.value_counts().head(10)

# Now arranging the data for T.V shows.
tv = file[file["type"] == "TV Show"]
listed_tv = pd.Series(tv["listed_in"])
category_tv = listed_tv.value_counts().head(10)

# For plotting the data.
plt.figure(figsize=(6, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', labeldistance = 1.03 ,radius = 1.2 ,startangle=140, textprops = {"fontsize":10})
plt.pie(category_tv, autopct = "%1.1f%%", labeldistance = 0.08 ,radius = 0.6, textprops = {"fontsize":8}, rotatelabels = True)

plt.title("10 POPULAR GENRE: MOVIES V/S T.V SHOWS", color = "r", fontsize = 20)
plt.legend(labels = category_tv.index, loc='upper left', bbox_to_anchor=(1.24,1.1), title = "T.V GENRES")
plt.show()
```

- `Dramas` dominate both movies (23.2%) and TV shows (28.4%), highlighting their universal popularity.
- `International Movies` and `International TV Shows` also have strong representation, indicating a global viewer base.

## Analysis Of Duration Of Movies:

```python
movie_d = file[file["type"] == "Movie"].copy()

# Convert duration for movies only
movie_d['duration_numeric'] = movie_d['duration'].str.replace(' min',"").astype(int)
sns.kdeplot(movie_d['duration_numeric'], fill = True, color = "blue")
plt.title("MOVIES DURATION", color = "r", fontsize = 20)
plt.xlabel("DURATION(minutes)",color = "#003366", fontsize = 15)
plt.ylabel("DENSITY",color = "#003366", fontsize = 15)
plt.show()
```

- Most movies have a duration around `90 to 110 minutes`, as shown by the tall central peak.

##  TV Shows With Largest Number Of Seasons:

```python
m = file[file["type"] == "TV Show"].copy()
m["tv_n"] = m["duration"].str.replace("Season","").str.replace("s","").str.strip().astype(int)
m_sorted = m.sort_values(by = "tv_n", ascending = False).head(50)
plt.figure(figsize = (12,6))
sns.barplot(x = "title",y = "tv_n", data = m_sorted, hue = "title", palette = "Set1")
plt.title("TV SHOWS WITH LARGEST SEASONS",color = "r", fontsize = 20)
plt.xlabel("T.V. SHOW NAMES",color = "#003366", fontsize = 15)
plt.ylabel("NO. OF SEASONS",color = "#003366", fontsize = 15)
plt.xticks(rotation = 90)
plt.show()
```

- `Supernatural` leads the chart with a massive `15 seasons`, making it the longest-running show in this dataset.

## Most Content Creating Countries:

```python
file1 = file["country"].value_counts().head(10).reset_index()
plt.figure(figsize = (10, 5))
sns.barplot(y = "country", x = "count", data = file1, hue = "country", palette = "viridis")
plt.title("10 POPULAR CONTENT CREATING COUNTRIES",color = "r", fontsize = 20)
plt.xlabel("NO. OF COUNTS",color = "#003366", fontsize = 15)
plt.ylabel("COUNTRIES",color = "#003366", fontsize = 15)
plt.show()
```

- `United States` & `India` are the top `Content Creating Countries`.
